```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
X_train = X_train.reshape(60000,784)
X_test = X_test.reshape(10000,784)
```

```python
#Prétraitement de données
X_train =X_train.astype('float32')
X_test=X_test.astype('float32') #des nombres virgules
X_train /= 255 #mettre les valeurs dans un intervalle [0,1]
X_test /= 255
#One_Hot Encoding
nb_classes = 10
y_Train = to_categorical(y_train, nb_classes) 
y_Test = to_categorical(y_test, nb_classes) 
```

### Création du modèle ###
Création du modèle de régression logistique. inputshape = 784 (28*28 pixels), image représentée sous forme de vecteur.

La fonction d'activation est la fonction softmax (qui est une généralisation de la fonction logistique). Elle représente dix classes, les dix chiffre de 0 à 9.

La fonction de perte est l'entropie croisée, la métrique utilisée pour mesurer la performance du modèle est l'accuracy (nombre de prédictions correctes sur le total de prédictions, le nombre d'images)

La sortie est constituée de 10 neurone

```python
model = Sequential()
model.add(Dense(10,input_shape =(784,),activation= 'softmax'))
model.compile(optimizer=SGD(learning_rate=0.05), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

Observer le nombres de paramètres à l'aide de la méthode summary(). Il y a 7850 = (784 + 1 (biais) x 10$ paramètres, 10 est le nombre de neurones de la couche de sortie.

```python
batch_size = 128
nb_epoch = 100
history = model.fit(X_train,y_Train,epochs=nb_epoch,batch_size=batch_size, verbose=1)
```

```python
evaluation = model.evaluate(X_test,y_Test,verbose =1)
print("Perte pour le jeu de test est %2f, Accuracy: %2f",evaluation[0],evaluation[1])
```
